```python
df = pd.read_csv('movie_review.csv')
df
```

J'extrait les colonnes dont j'ai besoin

```python
columns = df[['text', 'tag']]
columns
```

Je commence le pr√©traitement

```python
#Conversion du texte en minuscule
columns.loc[:, 'text'] = columns['text'].str.lower();
columns.loc[:, 'tag'] = columns['tag'].str.lower();
columns
```

```python
#Suppression des mots vides
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    words = text.split()
    filtered_text = [word for word in words if word not in stop_words]
    return ' '.join(filtered_text)
columns['text'] = columns['text'].apply(remove_stopwords)
columns['tag'] = columns['tag'].apply(remove_stopwords)
columns
```

```python
#Suppression de la ponctuation
import re
def remove_ponctuations(text):
  return re.sub(r'[^\w\s]', '', text)

columns['text'] = columns['text'].apply(remove_ponctuations)

columns

```

Entrainement du mod√®le Word2Vec

```python
from gensim.models import Word2Vec

phrases = [text.split() for text in columns['text']]
taille_vecteur = 150
fenetre = 7
model_word2vec = Word2Vec(phrases, vector_size=taille_vecteur, window=fenetre)
```

Vectorisation des reviews de movies

```python
import numpy as np
def calculer_review_embedding(review, model_word2vec, taille_vecteur):
    review_embedding = np.zeros(taille_vecteur)
    cp = 0
    for word in review.split():
        if word in model_word2vec.wv:
            review_embedding += model_word2vec.wv[word]
            cp +=1

    if cp != 0:

        review_embedding /= cp
    return review_embedding
```

Divisions des donn√©es

```python
from sklearn.model_selection import train_test_split

taille_vecteur = 150
X_train, X_test, y_train, y_test = train_test_split(columns['text'], columns['tag'], test_size=0.2, random_state=42)
X_train_embeddings = np.array([calculer_review_embedding(review, model_word2vec, taille_vecteur) for review in X_train])
X_test_embeddings = np.array([calculer_review_embedding(review, model_word2vec, taille_vecteur) for review in X_test])
```

Construction du classificateur LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_embeddings, y_train)
```

Evaluation du mod√®le

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
pred = model.predict(X_test_embeddings)
accuracy = accuracy_score(y_test, pred)
print("Accuracy :", accuracy)

# Calculer la pr√©cision
precision = precision_score(y_test, pred, average='weighted')
print("Pr√©cision :", precision)

# Calculer le rappel
recall = recall_score(y_test, pred, average='weighted')
print("Recall :", recall)

# Calculer le score F1
f1 = f1_score(y_test, pred, average='weighted')
print("F1 score :", f1)
```
