```python
import pandas as dp
from google.colab import files
import io
uploaded = files.upload()
```

```python
data = dp.read_csv(io.BytesIO(uploaded['sample.csv']))
data
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def clean_text(text):

    text = text.lower()

    text = re.sub(r'[^\w\s]', '', text)

    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

data['text_cleaned'] = data['text'].apply(clean_text)
data
```

```python
from collections import Counter

word_freq = Counter(' '.join(data['text_cleaned']).split())

num_top_words = 10
top_words = word_freq.most_common(num_top_words)

print("Mots les plus fréquents :", top_words)

most_common_words = set([word[0] for word in top_words])
data['text_cleaned'] = data['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in most_common_words]))

print(data.head())
```

```python
word_freq = {}
for text in data['text_cleaned']:
    words = text.split()
    for word in words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1


threshold = 2
rare_words = [word for word, freq in word_freq.items() if freq <= threshold]


data['text_cleaned'] = data['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in rare_words]))
print(data.head())
```

```python
data
```

```python
from nltk.stem import PorterStemmer
nltk.download('punkt')

stemmer = PorterStemmer()

data['text_stemmed'] = data['text_cleaned'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))
data
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

data['text_lemmatized'] = data['text_cleaned'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
data
```

```python
import re
from bs4 import BeautifulSoup
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1E0-\U0001F1FF"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)


def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

data['text_cleaned'] = data['text_cleaned'].apply(remove_emojis)


data['text_cleaned'] = data['text_cleaned'].apply(remove_urls)

data['text_cleaned'] = data['text_cleaned'].apply(remove_html_tags)

data
```
